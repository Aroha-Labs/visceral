```python
import pandas as pd
excel_file = 'Top_1000_data.xlsx'

# Load the specific sheet into a DataFrame
df = pd.read_excel(excel_file, sheet_name='Map Data')

# Display the first few rows of the DataFrame
df = df[["Game ","Description", "Age Rating", "Tags"]]
df
```

```python
final=df.drop_duplicates()
final
```

```python
final.nunique()
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(final, test_size=0.1, random_state=0)

train.head()
```

```python
train.to_csv('train.csv', index=False)
test.to_csv('test.csv', index=False)
```
